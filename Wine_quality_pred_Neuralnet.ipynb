##### This is a classification problem, label is Wine column and remaining are the feature columns. 
Assign feature and label columns

```python
X = df.drop('Wine', axis = 1) # drop the wine column from the dataset. This is features
y = df['Wine'] # this is label
```

```python
X.columns
```

```python
#get test train split
from sklearn.model_selection import train_test_split 
X_train, X_test, y_train, y_test = train_test_split(X,y) #By default it will assign 75 % of the data for training and 25 % for testing
```

```python
#Standardize the data: 
#We need to scale the data, due the large deviation in the values. 
from sklearn.preprocessing import StandardScaler 
scaler = StandardScaler()
```

```python
#Transform the train data
X_train = scaler.transform(X_train)
X_test = scaler.transform(X_test)
```

#### Neural network model using Multi-Layer Perceptron Classifier Model

```python
from sklearn.neural_network import MLPClassifier
```

```python
mlp = MLPClassifier(hidden_layer_sizes = (130,13,13), max_iter = 1500)
mlp.fit(X_train, y_train)
```

```python
#Predictions
pred = mlp.predict(X_test)
```

```python
 from sklearn.metrics import classification_report, confusion_matrix
```

```python
print(confusion_matrix(y_test, pred))
```

```python
print(classification_report(y_test, pred))
```
